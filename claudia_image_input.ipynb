```python
def sigmoid(x, L ,x0, k, b):
    y = L / (1 + np.exp(-k*(x-x0)))+b
    return (y)
```

```python
x_all = []
y_all = []
x_fit_all = []
y_fit_all = []

plt.rcParams.update({'figure.max_open_warning': 0})

for image,cond,selection,line_len in zip(image_names,condition,cut,lengths):
    data = io.imread(image)
    tresh = filters.threshold_li(data)
    #edge_sobel = filters.sobel(data)
    #data2 = data+2*edge_sobel
    #tresh = filters.threshold_li(data2)
    binary = data < tresh
    #plt.figure()
    #plt.imshow(edge_sobel, cmap="gray")
    #plt.figure()
    #plt.imshow(binary, cmap="gray")
    #print(image)
    plt.figure()
    plt.imshow(data, cmap="gray")
    #ax.set_title(image, fontsize=12)
    labels = skimage.measure.label(binary)
    seq = [region.area for region in measure.regionprops(labels)]
    max_index = seq.index(max(seq))
    my_label = labels==max_index+1
    closed_binary = morphology.closing(my_label, morphology.square(2))
    #print(my_label)
    image_label_overlay = color.label2rgb(closed_binary, image=data, bg_label=0)
    plt.figure()
    #plt.imshow(data, cmap="gray")
    plt.imshow(closed_binary)
    length_list=[]
    for column in closed_binary.T:
        length = np.count_nonzero(column)
        length_list.append(length)
    #cut movie
    length_list = length_list[selection[0]:selection[1]]
    # x axis in seconds
    duration_of_frame = 0.4
    time_in_frames = range(0,len(length_list))
    time_in_sec = [(frame * duration_of_frame)/3 for frame in range(0,len(length_list))]
    length_in_pixels = length_list
    #average 3 pixels
    N=3
    subList = [length_list[n:n+N] for n in range(0, len(length_list), N)]
    averaged_lengths = [mean(x)for x in subList]
    averaged_time_in_seconds = [frame * duration_of_frame for frame in range(0,len(subList))]
    # y axis in microns
    length_in_microns = [length * (line_len / data.shape[0]) for length in length_list]
    averaged_lengths_in_microns = [length * (line_len / data.shape[0]) for length in averaged_lengths]
    x_all.append(averaged_time_in_seconds)
    y_all.append(averaged_lengths_in_microns)
    plt.figure()
    plt.plot(time_in_sec, length_in_microns)
    plt.plot(averaged_time_in_seconds, averaged_lengths_in_microns)

    p0 = [max(averaged_lengths), np.median(averaged_time_in_seconds),1,min(averaged_lengths_in_microns)] # this is an mandatory initial guess
    try:
        popt, pcov = curve_fit(sigmoid, averaged_time_in_seconds, averaged_lengths_in_microns,p0, method='dogbox')
        x = averaged_time_in_seconds
        y = sigmoid(x, *popt)
        plt.plot(x,y, label='fit')
        x_fit_all.append(x)
        y_fit_all.append(y)
    except:
        print("Could not fit")
    #plt.plot(time_in_frames, length_in_microns)
    
    


```

```python

fig, axs = plt.subplots(5, figsize=(10,50))
for cond, x, y, x_fit, y_fit in zip(condition, x_all,y_all,x_fit_all, y_fit_all):
    if cond == "Tris":
        axs[0].set_title('Tris')
        axs[0].plot(x,y, "r")
        axs[0].plot(x_fit,y_fit, "r--")
    if cond == "Tris+GODCAT":
        axs[1].set_title(cond)
        axs[1].plot(x,y, "b")
        axs[1].plot(x_fit,y_fit, "b--")
    if cond == "Tris+BSA":
        axs[2].set_title(cond)
        axs[2].plot(x,y, "g")
        axs[2].plot(x_fit,y_fit, "g--")
    if cond == "Tris+BSA+DTT":
        axs[3].set_title(cond)
        axs[3].plot(x,y, "c")
        axs[3].plot(x_fit,y_fit, "c--")
    if cond == "Tris+DTT":
        axs[4].set_title(cond)
        axs[4].plot(x,y, "m")
        axs[4].plot(x_fit,y_fit, "m--")
for ax in axs:
    ax.set_ylabel('Diameter in microns')
    ax.set_xlabel('Time in seconds')

```

## Getting the length between the top and bottom edge of the segmentation mask
- This measures the inner part of the droplet. See the image above.
- It can also be easily modified to take the outer part.
- Or we can do it segmentation independent and use the two outer local Maxima
- The distance between the two droplets is not removed yet but can be easily done.

```python
length_list=[]
for column in my_label.T:
    top_pixel = None
    bottom_pixel = None
    for index, pixel in enumerate(column):
        if pixel == True:
            top_pixel = index
            break
    for index, pixel in enumerate(reversed(column)):
        if pixel == True:
            bottom_pixel = index
            break
    start = top_pixel
    stop = len(column) - bottom_pixel
    length = stop - start
    length_list.append(length)
```
